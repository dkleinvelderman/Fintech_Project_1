```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
csv_path = Path('../Resources/employment_illness.csv')
employmt_illness_data = pd.read_csv(csv_path)
employmt_illness_data.head(100)
```

```python
employmt_illness_data = employmt_illness_data.drop(columns=["Series ID", "Year", "Period"])
employmt_illness_data.head(29)
```

```python
employmt_illness_data = employmt_illness_data.rename(columns={"Label":"Date"})
employmt_illness_data.head(29)
```

```python
employmt_illness_data = employmt_illness_data[~employmt_illness_data['Date'].str.contains("Annual")]
```

```python
date = pd.to_datetime(employmt_illness_data['Date'], infer_datetime_format = True) #2020March and they are already strings
```

```python
#bring emp and date df together
date_df = pd.DataFrame(date)
date_df['Value'] = employmt_illness_data['Value']
date_df.hvplot.line("Date","Value", ylabel="# of illness in meat plants")
```

```python
date_df.hvplot.bar("Date","Value",ylabel = "# of illness in meat plants",rot=60) 
```
